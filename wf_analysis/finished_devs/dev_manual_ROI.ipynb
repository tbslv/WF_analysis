```python
import numpy as np

def separate_trials_by_stim(res_list, dffs, start=None, end=None):
    """
    Group trials by stimulus value (based on mean of stim[start:end])
    and compute mean DFF per stimulus.

    Parameters
    ----------
    res_list : list of dict
        Each element must contain a key 'stim' (stimulus trace, 1D array-like).
    dffs : list or array of np.ndarray
        List of DFF movies, each shaped (T, H, W).
    start : int or None
        Start index for the stimulus window (inclusive).
        If None, use the beginning of the stimulus trace.
    end : int or None
        End index for the stimulus window (exclusive).
        If None, use the end of the stimulus trace.

    Returns
    -------
    groups : dict
        Dictionary mapping each stimulus value (int) -> dict with:
            - 'indices': np.ndarray of trial indices
            - 'dff_stack': np.ndarray, shape (n_trials, T, H, W)
            - 'dff_mean': np.ndarray, shape (T, H, W)
    stimulus_order : np.ndarray
        Array of length n_trials with the stimulus value (int) for each trial.
    unique_stimuli : np.ndarray
        Sorted array of unique stimulus values.
    """

    # --------------------------
    # Compute stimulus_order like in your snippet
    # --------------------------
    if start is None:
        start = 0

    stimulus_order = np.array([
        int(tr['stim'][start:end].mean())  # same as your code, but res_list instead of files
        for tr in res_list
    ])

    unique_stimuli = np.unique(stimulus_order)

    # --------------------------
    # Group trials and collect DFFs
    # --------------------------
    groups = {}
    dff_shape = dffs[0].shape  # assume all same shape (T, H, W)

    for stimulus in unique_stimuli:
        # indices of trials with this stimulus
        trial_ids = np.where(stimulus_order == stimulus)[0]

        # stack DFFs for these trials
        dff_stack = np.zeros((len(trial_ids),) + dff_shape, dtype=dffs[0].dtype)
        for i, idx in enumerate(trial_ids):
            dff_stack[i] = dffs[idx]

        # compute mean across trials
        dff_mean = dff_stack.mean(axis=0)

        # store in dict
        groups[int(stimulus)] = {
            "indices": trial_ids,
            "dff_stack": dff_stack,
            "dff_mean": dff_mean,
        }

    return groups, stimulus_order, unique_stimuli
```
